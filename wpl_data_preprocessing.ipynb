```python
import pandas as pd
import json
```

# Processing Match Summary

```python
with open('RawData/wpl_matchSummary_2k23.json') as f:
    data = json.load(f)

df_matchSummary = pd.DataFrame(data)
df_matchSummary.head()
```

```python
df_matchSummary.shape
```

```python
df_matchSummary.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_matchSummary.head()
```

```python
match_ids = {}

for index, row in df_matchSummary.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids[key1] = row['match_id']
    match_ids[key2] = row['match_id']
```

```python
match_ids
```

```python
df_matchSummary.to_csv('ProcessedData/pre_prcs_match_summary.csv', index = False)
```

# Processing Batting Summary

```python
with open('RawData/wpl_battingSummary_2k23.json') as f:
    data = json.load(f)
  
df_battingSummary = pd.DataFrame(data)
df_battingSummary.head()
```

```python
def dismissal(dismissal):
    if dismissal == 'not out ' or dismissal == 'retired hurt ':
        return "not_out"
    else:
        return "out"

df_battingSummary['out/not_out'] = df_battingSummary.dismissal.apply(dismissal)
```

```python
df_battingSummary['match_id'] = df_battingSummary['match'].map(match_ids)
df_battingSummary.head(20)
```

```python
df_battingSummary.drop(columns=["dismissal"], inplace=True)
df_battingSummary.head(10)
```

```python
df_battingSummary.shape
```

```python
df_battingSummary.to_csv('ProcessedData/pre_prcs_batting_summary.csv', index = False)
```
