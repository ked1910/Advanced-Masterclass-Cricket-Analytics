Creating Phases of Over in Deliveries Dataframe

```python
def Phase (over):
  if over < 6:
    return 'Powerplay'
  elif over <= 15:
    return 'Middle'
  else:
    return 'Death'
```

```python
maindeliveriesdf['phase'] = maindeliveriesdf['over'].apply(lambda x : Phase(x))
maindeliveriesdf.tail()
```

Custom Functions

```python
#Strike Rate
def calcSR (runs,balls):
  return 100 * (runs / balls)
```

```python
#Runs per Innings
def calcRPI (runs,innings):
  return runs / innings
```

```python
#Balls per Dismissal
def calcBPD (balls,dismissals):
  if dismissals > 0:
    return balls/dismissals
  else:
    return balls/1
```

```python
#Balls per Boundary
def calcBPB (balls,boundary):
  if boundary > 0:
    return balls/boundary
  else:
    return balls/1
```

Main Function

```python
def CustomAnalysis (custDF,custVenue,custPhase,custOpposition):

  custDF = custDF[custDF.venue == custVenue]
  custDF = custDF[custDF.phase == custPhase]
  custDF = custDF[custDF.bowling_team == custOpposition]
  custDF.reset_index(inplace = True,drop = True)

  custDF['isDot'] = custDF['batsman_runs'].apply(lambda x: 1 if x == 0 else 0)
  custDF['isSingle'] = custDF['batsman_runs'].apply(lambda x: 1 if x == 1 else 0)
  custDF['isDouble'] = custDF['batsman_runs'].apply(lambda x: 1 if x == 2 else 0)
  custDF['isTriple'] = custDF['batsman_runs'].apply(lambda x: 1 if x == 3 else 0)
  custDF['isFour'] = custDF['batsman_runs'].apply(lambda x: 1 if x == 4 else 0)
  custDF['isSix'] = custDF['batsman_runs'].apply(lambda x: 1 if x == 6 else 0)

  Runs = pd.DataFrame(custDF.groupby(['batsman','matchId'])['batsman_runs'].sum().reset_index()).groupby(['batsman'])['batsman_runs'].sum().reset_index().rename(columns={'batsman_runs':'Runs'})
  Innings  = pd.DataFrame(custDF.groupby(['batsman'])['matchId'].apply(lambda x : len(list(np.unique(x)))).reset_index().rename(columns={'matchId':'Innings'}))
  Balls = pd.DataFrame(custDF.groupby(['batsman'])['matchId'].count().reset_index().rename(columns={'matchId':'Balls'}))
  Dismissals = pd.DataFrame(custDF.groupby(['batsman'])['player_dismissed'].count().reset_index().rename(columns={'player_dismissed':'Dismissals'}))

  Dots = pd.DataFrame(custDF.groupby(['batsman'])['isDot'].sum()).reset_index().rename(columns = {'isDot':'Dots'})
  Singles = pd.DataFrame(custDF.groupby(['batsman'])['isSingle'].sum()).reset_index().rename(columns = {'isSingle':'Singles'})
  Doubles = pd.DataFrame(custDF.groupby(['batsman'])['isDouble'].sum()).reset_index().rename(columns = {'isDouble':'Doubles'})
  Triples = pd.DataFrame(custDF.groupby(['batsman'])['isTriple'].sum()).reset_index().rename(columns = {'isTriple':'Triples'})
  Fours = pd.DataFrame(custDF.groupby(['batsman'])['isFour'].sum()).reset_index().rename(columns = {'isFour':'Fours'})
  Sixes = pd.DataFrame(custDF.groupby(['batsman'])['isSix'].sum()).reset_index().rename(columns = {'isSix':'Sixes'})

  newcustdF = pd.merge(Innings, Runs, on = 'batsman').merge(Balls, on = 'batsman').merge(Dismissals, on = 'batsman').merge(Dots, on = 'batsman').merge(Singles, on = 'batsman').merge(Doubles, on = 'batsman').merge(Triples, on = 'batsman').merge(Fours, on = 'batsman').merge(Sixes, on = 'batsman')

  newcustdF['SR'] = newcustdF.apply(lambda x : calcSR(x['Runs'], x['Balls']), axis = 1)
  newcustdF['RPI'] = newcustdF.apply(lambda x : calcRPI(x['Runs'], x['Innings']), axis = 1)
  newcustdF['BPB'] = newcustdF.apply(lambda x : calcBPB(x['Balls'], (x['Fours'] + x['Sixes'])), axis = 1)
  newcustdF['BPD'] = newcustdF.apply(lambda x : calcBPD(x['Balls'], x['Dismissals']), axis = 1)
  newcustdF['dot_percentage'] = newcustdF.apply(lambda x: x['Dots']/x['Balls'], axis = 1)


  return newcustdF
```

```python
myDF = CustomAnalysis(combineddF,'MA Chidambaram Stadium, Chepauk','Middle','Chennai Super Kings')
```

Calculating impacts score of the parameters with respect to weights

```python
myDF['cal_SR'] = myDF['SR'].apply(lambda x: x*x)
myDF['cal_RPI'] = myDF['RPI'].apply(lambda x: x*x)
myDF['cal_BPD'] = myDF['BPD'].apply(lambda x: x*x)
myDF['cal_dot_perc'] = myDF['dot_percentage'].apply(lambda x: x*x)

myDF.head(1)
```

```python
sq_sr, sq_rpi, sq_bpd, sq_dot_per = np.sqrt(myDF[['cal_SR','cal_RPI', 'cal_BPD', 'cal_dot_perc']].sum(axis = 0))
```

```python
sq_sr, sq_rpi, sq_bpd, sq_dot_per
```

```python
myDF['cal_SR'] = myDF['cal_SR'].apply(lambda x: x/sq_sr)
myDF['cal_RPI'] = myDF['cal_RPI'].apply(lambda x: x/sq_rpi)
myDF['cal_BPD'] = myDF['cal_BPD'].apply(lambda x: x/sq_bpd)
myDF['cal_dot_perc'] = myDF['cal_dot_perc'].apply(lambda x: x/sq_dot_per)

myDF.head(1)
```

```python
myDF['cal_SR'] = myDF['cal_SR'].apply(lambda x: x*wsr)
myDF['cal_RPI'] = myDF['cal_RPI'].apply(lambda x: x*wrpi)
myDF['cal_BPD'] = myDF['cal_BPD'].apply(lambda x: x*wbpd)
myDF['cal_dot_perc'] = myDF['cal_dot_perc'].apply(lambda x: x*wdper)

myDF.head(1)
```

```python
best_sr, worst_sr = max(myDF['cal_SR']), min(myDF['cal_SR'])
best_rpi, worst_rpi = max(myDF['cal_RPI']), min(myDF['cal_RPI'])
best_bpd, worst_bpd = max(myDF['cal_BPD']), min(myDF['cal_BPD'])
best_dot_perc, worst_dot_perc = min(myDF['cal_dot_perc']), max(myDF['cal_dot_perc'])
```

```python
best_sr, worst_sr,best_rpi, worst_rpi,best_bpd, worst_bpd,best_dot_perc, worst_dot_perc
```

```python
myDF['my_best_SR'] = myDF['cal_SR'].apply(lambda x: (x-best_sr)*(x-best_sr))
myDF['my_best_RPI'] = myDF['cal_RPI'].apply(lambda x: (x-best_rpi)*(x-best_rpi))
myDF['my_best_BPD'] = myDF['cal_BPD'].apply(lambda x: (x-best_bpd)*(x-best_bpd))
myDF['my_best_dot_perc'] = myDF['cal_dot_perc'].apply(lambda x: (x-best_dot_perc)*(x-best_dot_perc))

myDF.head(1)
```

```python
myDF['my_best_sqrt'] = myDF.apply(lambda x: x['my_best_SR'] + x['my_best_RPI'] + x['my_best_BPD'] + x['my_best_dot_perc'], axis = 1)

myDF.head(1)
```

```python
myDF['my_worst_SR'] = myDF['cal_SR'].apply(lambda x: (x-worst_sr)*(x-worst_sr))
myDF['my_worst_RPI'] = myDF['cal_RPI'].apply(lambda x: (x-worst_rpi)*(x-worst_rpi))
myDF['my_worst_BPD'] = myDF['cal_BPD'].apply(lambda x: (x-worst_bpd)*(x-worst_bpd))
myDF['my_worst_dot_perc'] = myDF['cal_dot_perc'].apply(lambda x: (x-worst_dot_perc)*(x-worst_dot_perc))

myDF.head(1)
```

```python
myDF['my_worst_sqrt'] = myDF.apply(lambda x: x['my_worst_SR'] + x['my_worst_RPI'] + x['my_worst_BPD'] + x['my_worst_dot_perc'], axis = 1)

myDF.head(1)
```

```python
myDF['impactScore'] = myDF.apply(lambda x: x['my_worst_sqrt']/(x['my_worst_sqrt'] + x['my_best_sqrt']), axis = 1)

myDF.head()
```
